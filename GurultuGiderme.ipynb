```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
resim1= cv2.imread("/content/gdrive/MyDrive/median.jpg")
```

```python
resim2= cv2.medianBlur(resim1,9)  #9, blurlama buyuklugudur
```

```python
plt.subplot(1,2,1), plt.imshow(resim1), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(resim2), plt.title("median")
plt.xticks([]), plt.yticks([])
plt.show
```

```python
resim_bilateral= cv2.imread("/content/gdrive/MyDrive/bilateral.png")
```

```python
resim3= cv2.bilateralFilter(resim_bilateral,9,100,100)
```

```python
plt.subplot(1,2,1), plt.imshow(resim_bilateral), plt.title("orijinal")
plt.xticks([]), plt.yticks()

plt.subplot(1,2,2), plt.imshow(resim3), plt.title("bilateral filtre")
plt.xticks([]), plt.yticks([])

plt.show()  #icindeki ve disindaki cizgiler kayboldu ama sinirari net bir sekilde gorebiliyoruz.
```
